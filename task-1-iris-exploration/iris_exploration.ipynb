```python
# Step 1: Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the Iris Dataset
iris = sns.load_dataset('iris') 
```

```python
# Step 3: Inspect the Dataset
print("Shape of the dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst five rows:")
print(iris.head())
```

```python
# Step 4: Dataset Info and Summary Statistics
print("\nDataset Info:")
print(iris.info())
```

```python
print("\nSummary Statistics:")
print(iris.describe())
```

```python
# Step 5: Scatter Plot (Petal Length vs Petal Width)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# Step 6: Histograms for Feature Distributions
iris.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle('Feature Distributions', fontsize=16)
plt.show()
```

```python
# Step 7: Box Plots to Detect Outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title('Box Plot of Iris Features')
plt.show()
```

```python
 #Pair Plot for Overall Visualization
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```
